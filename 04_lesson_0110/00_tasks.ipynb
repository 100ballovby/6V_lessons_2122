<p><b>Напишите программу, которая считывает три числа и выводит их сумму. Каждое число записано в отдельной строке.</b></p>

<p>Для решени задачи нам понадобится 3 переменные.</p>

```python
a = int(input('Введите число: '))  # буду хранить тут числа
b = int(input('Введите число: '))  # буду хранить тут числа
c = int(input('Введите число: '))  # буду хранить тут числа

print(f'{a} ⭐️ {b} ⭐️ {c} = {a + b + c}')
```

<p>Буква <code>f</code> перед строкой позволяет вставить в строку значение переменной в <code>{}</code>. Это не несет никакой смысловой нагрузки, просто красивый способ вывода. Обратите внимание, что символы + работают как операция только в самой последней фигурной скобочке.</p>

<p><b>Дано число <code>n</code>. С начала суток прошло <code>n</code> минут. Определите, сколько часов и минут будут показывать электронные часы в этот момент. Программа должна вывести два числа: количество часов (от <code>0</code> до <code>23</code>) и количество минут (от <code>0</code> до <code>59</code>). Учтите, что число <code>n</code> может быть больше, чем количество минут в сутках.</b></p>

<p>Для решения задачи мне понадобится 3 переменные. <code>n</code> - это ввод пользователя. <code>hours</code>, в которой я буду считать количество часов, переменная <code>minutes</code>, в которой я буду считать количество минут.</p>

```python
'''Электронные часы'''
n = int(input('Сколько минут прошло? '))
hours = n % (60 * 24) // 60
minutes = n % 60

print(f'{hours} ч., {minutes} мин.')
```

<ul>
    <li>Формула <code>n % (60 * 24) // 60</code> находит количество часов. Так как количество минут может быть больше, чем есть в сутках, я сначала нахожу количество минут в сутках <code>(60 * 24) = 1440</code>, а затем нахожу остаток от деления <code>n % 1440</code> и делю его на 60. Таким образом я получаю количество часов.</li>
    <li>Формула <code>n % 60</code> считает количество минут. Если мне введут число <code>189</code>, я найду остаток от деления на 60, который будет равен 9. 189 - это 3 часа и 9 минут.</li>
</ul>
